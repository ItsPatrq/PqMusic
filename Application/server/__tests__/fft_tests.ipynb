```python
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.fftpack import fft, ifft

Fs = 44100.0
T = 1.0 / Fs
f = 440
sample = 8000

x = np.arange(sample)
y = np.sin(2 * np.pi * f * x / Fs)
secound_sine = np.sin(2 * np.pi * 7800 * x / Fs)
y += secound_sine
plt.figure(0)
plt.plot(x, y)
plt.xlabel('sample(n)')
plt.ylabel('voltage(V)')
plt.show()
print(len(y))
outfft = fft(y)
fft_len = int(np.floor(len(outfft) / 2))
powerfft = np.log(abs(outfft))
print(len(powerfft))
xf = np.linspace(0, 1.0 / (2.0 * T), fft_len)
plt.plot(xf, 2.0 / sample *  powerfft[:fft_len])
plt.xlabel('hz')
plt.ylabel('')
```
